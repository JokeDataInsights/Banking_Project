```python
def categorize_loan_term(term):
    if term in [12, 24]:
        return 'Short Term'
    elif term == 36:
        return 'Medium Term'
    elif term in [48, 60]:
        return 'Long Term'
    else:
        return'Other'
```

```python
# Apply category
df['Loan Term Category'] = df['Loan Term'].apply(categorize_loan_term)


```

```python
# Anomaly Detection
#  Which transactions or records are flagged as anomalies?

anomalies = df[df['Anomaly'] == True] 
print(anomalies)
```

```python
# Ensuring datetime is in the correct format
anomalies['Transaction Date'] = pd.to_datetime(anomalies['Transaction Date'])

# Create Day of Week column
anomalies['Day of Week'] = anomalies['Transaction Date'].dt.day_name()

# Create pivot table
heat_data = anomalies.pivot_table(index='Day of Week', columns='Transaction Type', aggfunc='size', fill_value=0)

# Reorder days
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heat_data = heat_data.reindex(days_order)



# Plotting heatmap
plt.figure(figsize=(10,6))
sns.heatmap(heat_data, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Heatmap of Anomalies by Day and Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Day of Week')
plt.tight_layout()
plt.show()

```

```python
# Ensuring the datetime is in the right format
anomalies['Transaction Date'] = pd.to_datetime(anomalies['Transaction Date'])

# Extracting month and day
anomalies['Month'] = anomalies['Transaction Date'].dt.strftime('%B')

anomalies['Day'] = anomalies['Transaction Date'].dt.day

# Pivot table
month_day_data = anomalies.pivot_table(index='Day', columns='Month', aggfunc='size', fill_value=0)

# Sorting  months in calendar order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
month_day_data = month_day_data.reindex(columns=month_order, fill_value=0)

# Plotting heatmap
plt.figure(figsize=(14, 7))
sns.heatmap(month_day_data, cmap='Purples', linewidths=0.5, annot=True, fmt='d')
plt.title('Heatmap of Anomalies by Month and Day')
plt.xlabel('Month')
plt.ylabel('Day of Month')
plt.tight_layout()
plt.show()
```
